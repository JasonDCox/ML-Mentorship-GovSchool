```python
model_name = "tfLite" #tfLite, tensorRT, or Tensorflow
csv_location = model_name + "\\" + model_name + "_results.csv"
df = pd.read_csv(csv_location)
```

```python
offset = 0
if model_name == "tensorRT":
    offset = -1


start = model_name + "\\predictions"


#Create text file for predictions on each image from the data frame
empties = 0
for i in range(df.shape[0]):
    requirement = 1
    limit = 1
    threshold = 0.0
    counter = 0
    row = df.values.tolist()[i]
    name = row[1]
    file_name = name[0:-3] + "txt"
    file_path = start + "\\" + file_name
    things = row[4:]
    width = int(row[2])
    height = int(row[3])
    #print(things)
    with open(file_path, "w") as write_file:
        for x in things:
            if str(x) != "nan":
                #print(x)
                coordinates = x.split("[")[1].split(']')[0].strip().replace(" ", "").split(",")
                confidence = float(x.split(",")[6].split(":")[1].replace("'", "").replace("}","").strip())
                guess_id = x.split(",")[5].split(":")[1].strip()
                #print(int(float(guess_id)))
                guess = labels[int(float(guess_id)) + offset]["name"]
                #print(guess)
                line = guess + " " + str(confidence) + " " +  str(int(float(coordinates[1]) * width)) + " " + str(int(float(coordinates[0]) * height)) + " " + str(int(float(coordinates[3]) * width)) + " " + str(int(float(coordinates[2]) * height))
                if counter < requirement or confidence > threshold:
                    #print(line)
                    write_file.write(line + "\n")
                    counter += 1
    if counter == 0:
        empties += 1
    #print(i)
print("empties:", empties)
```

```python
#convert avi videos to mp4 if needed

import cv2
import numpy as np
import time

vid_url = model_name + "\\" + model_name + ".avi"
cap = cv2.VideoCapture(vid_url)

# Check if camera opened successfully
if (cap.isOpened()== False):
    print("Error opening video stream or file")

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
print("")

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('converted_vid.mp4',fourcc, 10, (frame_width,frame_height))
# Read until video is completed
while(cap.isOpened()):
    # Capture frame-by-frame
    ret, frame = cap.read()
    if ret == True:
        
        out.write(frame)
        
        # Display the resulting frame
        cv2.imshow('Frame',frame)
        
        # Press Q on keyboard to  exit
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    # Break the loop
    else:
        break

cap.release()
out.release()

# Closes all the frames
cv2.destroyAllWindows()
```
